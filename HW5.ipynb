```python
# Q2, part 1
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
import pymc
import arviz as az

np.random.seed(365)
n = 30
true_theta, true_tau = -2, 1/36
x = stats.norm(loc = -2, scale = 6).rvs(n)
conjugate_normal_gamma = pymc.Model()
with conjugate_normal_gamma: #open a resource and guarantee that the resource will be closed when the with block completes, regardless of how the block completes.

    # Priors for unknown model parameters
    theta0 = 0  # prior belief regarding true_theta
    theta_prior_n = 5  # strength of prior belief as units of data
    # theta_prior_n = tau0/true_tau
    tau0 = theta_prior_n*true_tau  
    theta = pymc.Normal("theta", mu=theta0, sigma=tau0**(-0.5))
    # Compared to scipy.stats
    # loc -> mu
    # scale -> sigma but you can also use tau as below...

    tau_SS_prior = 8  # prior belief regarding true_sum_of_squares
    tau_prior_n = 0.5  # strength of prior belief as units of data
    tau = pymc.Gamma("tau", alpha=tau_prior_n/2, 
                                beta=tau_SS_prior/2)

    # Likelihood (sampling distribution) of observations
    x_obs = pymc.Normal("x_obs", mu=theta, tau=tau, observed=x)

m = 10000
with conjugate_normal_gamma:
    # draw m posterior samples
    idata = pymc.sample(draws=m, chains=2, tune=100)
    # arviz.InferenceData object
# automatically tunes params (gives a good acceptance rate like 80%)

theta = idata.posterior['theta'].values
tau = idata.posterior['tau'].values
# bayesian inference
summ = az.summary(idata, round_to=5)
print("mean of the data: ", np.mean(x))
print("PyMC estimated mean (3%, 97% hdi): ", summ["mean"]["theta"], 
      " (", summ["hdi_3%"]["theta"],", ", summ["hdi_97%"]["theta"], ")")
print("precision of the data: ", 1/np.var(x))
print("PyMC estimated precision (3%, 97% hdi): ", summ["mean"]["tau"], 
      " (",summ["hdi_3%"]["tau"] ,", ", summ["hdi_97%"]["tau"], ")")
# diagnostics
az.plot_forest(idata, var_names=["theta", "tau"], 
               combined=True, hdi_prob=0.95, r_hat=True);
# combined = True means we are combining chains, and r_hat now shows up
# r_hat measures how different the 2 chains are (1 means there is no difference between chains)
fig,ax = plt.subplots(1,2,figsize=(10,4))
ax[0].plot(theta.ravel()[:-1], theta.ravel()[1:], '.', alpha=0.25)
ax[1].plot(tau.ravel()[:-1], tau.ravel()[1:], '.', alpha=0.25);
ax[0].set_title("Autocorrelation visualization for sampled thetas")
ax[1].set_title("Autocorrelation visualization for sampled taus")
```

```python
# Q2, part 2
n = 30
true_theta, true_tau = -2, 1/36
x = stats.norm(loc = -2, scale = 6).rvs(n)
moyal_pareto = pymc.Model()
with moyal_pareto: #open a resource and guarantee that the resource will be closed when the with block completes, regardless of how the block completes.

    # Priors for unknown model parameters
    theta0 = 0  # prior belief regarding true_theta
    theta_prior_n = 5  # strength of prior belief as units of data
    # theta_prior_n = tau0/true_tau
    tau0 = theta_prior_n*true_tau  
    theta = pymc.Moyal("theta", mu=theta0, sigma=tau0**(-0.5))
    # Compared to scipy.stats
    # loc -> mu
    # scale -> sigma but you can also use tau as below...

    tau_SS_prior = 1  # prior belief regarding true_sum_of_squares
    tau_prior_n = 0.5  # strength of prior belief as units of data
    tau = pymc.Pareto("tau", m=tau_SS_prior, 
                                alpha=1/tau_prior_n)

    # Likelihood (sampling distribution) of observations
    x_obs = pymc.Normal("x_obs", mu=theta, tau=tau, observed=x)

m = 10000
with moyal_pareto:
    # draw m posterior samples
    idata = pymc.sample(draws=m, chains=2, tune=100)
    # arviz.InferenceData object
# automatically tunes params (gives a good acceptance rate like 80%)

theta = idata.posterior['theta'].values
tau = idata.posterior['tau'].values
# bayesian inference
summ = az.summary(idata, round_to=5)
print("mean of the data: ", np.mean(x))
print("PyMC estimated mean (3%, 97% hdi): ", summ["mean"]["theta"], 
      " (", summ["hdi_3%"]["theta"],", ", summ["hdi_97%"]["theta"], "), using Moyal(mu=theta0, sigma=tau0^(-0.5)) prior.")
print("precision of the data: ", 1/np.var(x))
print("PyMC estimated precision (3%, 97% hdi): ", summ["mean"]["tau"], 
      " (",summ["hdi_3%"]["tau"] ,", ", summ["hdi_97%"]["tau"], "), using Pareto(m=tau_SS_prior, alpha=1/tau_prior_n) prior.")
print("PyMC isn't magical. We started with a bad starting parameterization where the support of the Pareto distribution we're sampling to get taus doesn't include the tau of our data. The sampler tries to get to 0.03..., as reflected in the very narrow high density credible interval that includes the minimum value of distribution's support, but it simply can't.")
# diagnostics
az.plot_forest(idata, var_names=["theta", "tau"], 
               combined=True, hdi_prob=0.95, r_hat=True);
# combined = True means we are combining chains, and r_hat now shows up
# r_hat measures how different the 2 chains are (1 means there is no difference between chains)
fig,ax = plt.subplots(1,2,figsize=(10,4))
ax[0].plot(theta.ravel()[:-1], theta.ravel()[1:], '.', alpha=0.25)
ax[1].plot(tau.ravel()[:-1], tau.ravel()[1:], '.', alpha=0.25);
ax[0].set_title("Autocorrelation visualization for sampled thetas")
ax[1].set_title("Autocorrelation visualization for sampled taus")
```

### Q3: Slice Sampling

*First explain how the Markov algorithm of slice sampling as given below works. Then explain the steps by which slice sampling could be used in place of a Metropolis-Hasting step in a Metropolis within Gibbs algorithm where the full conditionals are only known up to a normalizing constant. In your explanation clarify what the curve that we're sampling beneath is, and what the initial value and steps are to create the draw for Gibbs sampling.*

1. We have m samples and give a starting x-coordinate ($x_1$) within the support of our pdf.
2. Then we get a random number from 0 to $pdf(x_1)$, $y_1$. The range of possible values for $y_1$ is proportional to the likelihood that $x_1$ is produced by our target distribution.
3. We call slice_f_at_y to get $x_2$. This is done by recursively narrowing an interval [a,b] such that when we get a random number from [a,b], $x'$, $pdf(x') > y_1$ and we can let $x_2 = x'$.
4. Repeat steps 2 and 3 for m iterations to get m samples, which are the $x$'s we get from slice_f_at_y.

Intuitively, this procedure works because we require $pdf(x_{next}) > y_{current}$ and $x$'s with higher density will satisfy this condition proportionally more than $x$'s with lower density in our target distribution.

To replace Metropolis-Hasting (MH) in a Metropolis within Gibbs algorithm, we can start with with initializing our first $x_1$ as a real number within the support of our "pdf", which is the full conditional distribution. Then instead of following the MH algorithm of sampling from a proposal distrbution and deciding whether to reject this sample or not, we get our next samples from following steps 2-4 from above where we sample from under the full conditional distribution. We accept a new sample value every time we use slice sampling, unlike MH where we may reject values. Since slice sampling is only comparing values which are generated from under the same full conditional distribution, everything is proportional and normalizing constants are irrelevant.

```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np

def slice_f_at_y(f, x, y, x_grid=np.linspace(0,1,51)):
    
    # find interval of grid points where f(x_grid) > y 
    # then extend the enterval so endpoints f(a)<y and f(b)<y 
    x_grid_delta = x_grid[1]-x_grid[0]
    a,b = x_grid[f(x_grid)>y][[0,-1]]+[-x_grid_delta,x_grid_delta]
    # a,b = x_grid[0,-1]  # make the interval all of x_grid
    
    x_ = a + stats.uniform().rvs()*(b-a)  # get random x coord in [a,b], which is contained in [0,y] s.t. f(a)<y and f(b)<y 
    if f(x_)>y:
        return x_,1  # in 1 try if f(x_)>y and "x_ is under f" (i.e x_ in horizontal slice at y = y under pdf)
    elif x_ < x:  # or if "x_ was above f on the left side of the interval" 
        x_l,x_r = x_,b # fine tunes left end of interval (makes it a tiny bit bigger)
    else:
        x_l,x_r = a,x_  # or if "x_ was above f on the right side of the interval"
        # fine tunes right end of interval (makes it a tiny bit smaller)
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = 2)  # try again with a reduced interval

def slice_f_at_y_(f, x, y, x_l=0, x_r=1, tot=1):
    
    x_ = x_l + stats.uniform().rvs()*(x_r-x_l)
    
    if f(x_)>y:
        return x_,tot
    elif x_ < x:
        x_l = x_
    else:
        x_r = x_
    return slice_f_at_y_(f, x, y, x_l, x_r, tot = tot+1) 
    # recursively fine tune interval bounds until we get an x_ in horizontal slice at y = y under pdf
```

```python
x_grid = np.linspace(0,1,1000)
f = lambda x: stats.beta(1.5,3).pdf(x)
plt.plot(x_grid, f(x_grid))
    
m = 1000
x = np.zeros([m+1,3])
x[:2,0] = 0.25

plot_trace = 10
for t in range(1,m):
    
    x[t,1] = stats.uniform().rvs()*f(x[t,0]) #number from 0 to f(x[t,0])
    if t < plot_trace:
        plt.plot([x[t,0]]*2, [x[t-1,1],x[t,1]], 'k')
    
    x[t+1,0],x[t+1,2] = slice_f_at_y(f, x[t,0], x[t,1]) # pdf, x coord, y coord <= density at x coord
    if t < plot_trace:
        if t==1:
            plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--', label=str(plot_trace)+ " iterations")
        plt.plot([x[t,0], x[t+1,0]], [x[t,1]]*2, 'k--')
    
plt.hist(x[:,0], density=True, label=str(m)+" iterations\n x values only")
plt.legend();
```
